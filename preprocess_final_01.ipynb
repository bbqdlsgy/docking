# 2. 리간드 분자량 250~400인 pdb만 추출한 csv 가져오기

```python
import pandas as pd
import numpy as np
import pickle

from tqdm import tqdm
import logging

import pubchempy as pcp
from pubchempy import Compound, get_compounds, get
```

```python
pdb_lig = pd.read_csv('pdb_2023_human_X_ray_Prot_resol2/01_preprocessed_pdb.csv')
pdb_lig.info()
```

```python
print('Ligand ID:', len(pdb_lig['Ligand ID'].unique()), '\n',
      'InChI: ', len(pdb_lig['InChI'].unique()), '\n',
      'InChI Key: ', len(pdb_lig['InChI Key'].unique()), '\n',
      'Ligand Name: ', len(pdb_lig['Ligand Name'].unique()))
```

### ligand ID, ligand SMILES 다 시도해보았지만 완전히 못 찾기 때문에 cid,isomeric_smiles, iupac_name을 None값으로 저장함.

```python
# pdb_lig['isomeric_smiles']=""
# pdb_lig['CID'] = ""
# pdb_lig['iupac_name'] = ""
# i = 0

isosmi = []
cids = []
iupac_names = []
r = 0
for inchi in tqdm(pdb_lig['InChI']):
      try:
            compound = get_compounds(inchi, namespace='inchi')
            if (compound[0].cid is None):
                  print("Since inch cannot find a compound, we will search by ligand name.", '\n',
                        inchi, compound, '\n',
                        'ligand name: ', pdb_lig['Ligand Name'][r])
                  compound = get_compounds(pdb_lig['Ligand Name'][r], namespace='name')
                  isomeric_smiles, cid, iupac_name = compound[0].isomeric_smiles, compound[0].cid, compound[0].iupac_name
                  isosmi.append(isomeric_smiles)
                  cids.append(cid)
                  iupac_names.append(iupac_name)
                  print('row: ', r+2, 'cid: ', compound[0].cid, '\n', '#'*150) # r+2는 그냥 체크하기 쉽게하기 위함.
                  r += 1
            else:
                  isomeric_smiles, cid, iupac_name = compound[0].isomeric_smiles, compound[0].cid, compound[0].iupac_name
                  r += 1
                  
                  isosmi.append(isomeric_smiles)
                  cids.append(cid)
                  iupac_names.append(iupac_name)
            
            # Ignore SettingWithCopy Warning 
            # pd.set_option('mode.chained_assignment', None)
            
            # make new column & insert values
            # pdb_lig['isomeric_smiles'][i] = isomeric_smiles
            # pdb_lig['CID'][i] = cid
            # pdb_lig['iupac_name'][i] = iupac_name
            # i += 1
            
      except IndexError as ie:
            print(r+2, 'IndexError: ', ie) # r+2는 그냥 체크하기 쉽게하기 위함.
            r += 1
            isosmi.append(None)
            cids.append(None)
            iupac_names.append(None)
            
            # isosmi.append('0')
            # cids.append(0)
            # iupac_names.append('0')
      except TypeError as t:
            print(r+2, 'TypeError: ',t) # r+2는 그냥 체크하기 쉽게하기 위함.
            r += 1
            isosmi.append(None)
            cids.append(None)
            iupac_names.append(None)
            
            # isosmi.append('0')
            # cids.append(0)
            # iupac_names.append('0')
            
with open('pdb_2023_human_X_ray_Prot_resol2/pubchem_isosmi.pickle', 'wb') as pi:
      pickle.dump(isosmi, pi)
      
with open('pdb_2023_human_X_ray_Prot_resol2/pubchem_cids.pickle', 'wb') as pc:
      pickle.dump(cids, pc)
      
with open('pdb_2023_human_X_ray_Prot_resol2/pubchem_iupac_names.pickle', 'wb') as pn:
      pickle.dump(iupac_names, pn)
```

```python
with open('pdb_2023_human_X_ray_Prot_resol2/pubchem_isosmi.pickle', 'rb') as pi:
      isosmi = pickle.load(pi)
      
with open('pdb_2023_human_X_ray_Prot_resol2/pubchem_cids.pickle', 'rb') as pc:
      cids = pickle.load(pc)
      
with open('pdb_2023_human_X_ray_Prot_resol2/pubchem_iupac_names.pickle', 'rb') as pn:
      iupac_names = pickle.load(pn)
```

```python
pdb_lig['isomeric_smiles']=""
pdb_lig['CID'] = ""
pdb_lig['iupac_name'] = ""
i = 0

for isomeric_smiles, cid, iupac_name in zip(isosmi, cids, iupac_names):
    # make new column & insert values
    pdb_lig['isomeric_smiles'][i] = isomeric_smiles
    pdb_lig['CID'][i] = cid
    pdb_lig['iupac_name'][i] = iupac_name
    i += 1
```

```python
pdb_lig.to_csv('pdb_2023_human_X_ray_Prot_resol2/02_final_preprocessed_pdb.csv', index=False, index_label=False)
```

# 5. Resolution 최소인 것 추출하기

```python
path = 'pdb_2023_human_X_ray_Prot_resol2/final_output/pubchem_3d_exist/final_rcsb_pubchempy_2023_3d_exist.csv'
final_pdb_lig_prot_info_3d_exist = pd.read_csv(path)


path = 'pdb_2023_human_X_ray_Prot_resol2/final_output/pubchem_3d_empty/final_pdb_lig_prot_info_3d_empty.csv'
final_pdb_lig_prot_info_3d_empty = pd.read_csv(path)

print('Length of 3d_exist:', len(final_pdb_lig_prot_info_3d_exist), '\n',
      'Length of 3d_empty:',len(final_pdb_lig_prot_info_3d_empty))
```

```python
exist_authors = final_pdb_lig_prot_info_3d_exist['Structure Author'].unique().tolist()

exist_resol = dict()
for author in exist_authors:
    author_df = final_pdb_lig_prot_info_3d_exist[final_pdb_lig_prot_info_3d_exist['Structure Author']==author]
    if author not in exist_resol.keys():
        exist_resol[author]= min(author_df['Resolution (Å)'])
    else:
        continue
    
print('Number of exist_Author: ', len(final_pdb_lig_prot_info_3d_exist['Structure Author'].unique()))

empty_authors = final_pdb_lig_prot_info_3d_empty['Structure Author'].unique().tolist()

empty_resol = dict()
for author in empty_authors:
    author_df = final_pdb_lig_prot_info_3d_empty[final_pdb_lig_prot_info_3d_empty['Structure Author']==author]
    if author not in empty_resol.keys():
        empty_resol[author]= min(author_df['Resolution (Å)'])
    else:
        continue
    
print('Number of empty_Author: ', len(final_pdb_lig_prot_info_3d_empty['Structure Author'].unique()))
```

# 3d 여러개 다운로드할 때

```python
# pdb, cid, lid of sdf
exist_3d_pdb = sdf['PDB'].tolist()
exist_3d_cid = sdf['CID'].tolist()
exist_3d_lid = sdf['LID'].tolist()
exist_3d_isosmi = sdf['isomeric_smiles'].tolist()

# original pdb path
ori_pdb_path = 'pdb_2023_human_X_ray_Prot_resol2/rcsb_api_pdb'
ori_pdb = os.listdir(ori_pdb_path)

# save pdb & 3d conformer
output = 'pdb_2023_human_X_ray_Prot_resol2/final_output/pubchem_3d_exist/pdb_lig_multiple'

for digit, cid, lid, iso_smi in zip(exist_3d_pdb, exist_3d_cid, exist_3d_lid, exist_3d_isosmi):
    os.makedirs(output + f"/{digit}", exist_ok=True)

    try:
        id_url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/{cid}/conformers/TXT?record_type=3d"
        response = urllib.request.urlopen(id_url)
        saveid = f"{output}/{digit}/{digit}_{lid}_{cid}_3d_id.txt"
        urllib.request.urlretrieve(id_url, saveid)
        
        response = requests.get(id_url)
        for line in response.text.splitlines():
            print(f"{line}_{cid}")
            id_cid_url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/conformers/{line}/SDF?response_type=save&response_basename=Conformer3D_COMPOUND_CID_{cid}_{line}"
            response = urllib.request.urlopen(id_cid_url)
            saveid = f"{output}/{digit}/Conformer3D_COMPOUND_{digit}_{lid}_{cid}_{line}.sdf"
            urllib.request.urlretrieve(id_cid_url, saveid)
    except:
        print('except')
```

```python
# pdb, cid, lid of sdf
exist_3d_pdb = sdf['PDB'].tolist()
exist_3d_cid = sdf['CID'].tolist()
exist_3d_lid = sdf['LID'].tolist()
exist_3d_isosmi = sdf['isomeric_smiles'].tolist()

# original pdb path
ori_pdb_path = 'pdb_2023_human_X_ray_Prot_resol2/rcsb_api_pdb'
ori_pdb = os.listdir(ori_pdb_path)

# save pdb & 3d conformer
output = 'pdb_2023_human_X_ray_Prot_resol2/final_output/pubchem_3d_exist/pdb_lig_single'

# check = []
for digit, cid, lid, iso_smi in zip(exist_3d_pdb, exist_3d_cid, exist_3d_lid, exist_3d_isosmi):
    os.makedirs(output + f"/{digit}", exist_ok=True)
    
    pdb = digit + '.pdb'
    if pdb in ori_pdb:
        # check.append(pdb)
        sp = subprocess.Popen(f"cp -v {ori_pdb_path}/{pdb} {output}/{digit}/{digit}_{lid}_{cid}.pdb", 
                        shell=True, stdout=subprocess.PIPE,  
                        stderr=subprocess.PIPE)
    
        s, e = sp.communicate()
        print(s.decode('utf-8'))
        print(e.decode('utf-8'))
        print(sp.returncode)
    try:
        url = f'https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/smiles/{iso_smi}/SDF?record_type=3d&response_type=save&response_basename=Conformer3D_COMPOUND_CID_{cid}'
        response = urllib.request.urlopen(url)
        savename = f"{output}/{digit}/Conformer3D_COMPOUND_{digit}_{lid}_{cid}.sdf"
        
        # url이 가리키는 주소에 접근해서 해당 자원을 로컬 컴퓨터에 저장하기
        urllib.request.urlretrieve(url, savename)
    except HTTPError as herr:
        print(f'{digit}_{lid}_{cid} ' + 'Reason: ', herr.reason)
        url = f'https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/{cid}/SDF?record_type=3d&response_type=save&response_basename=Conformer3D_COMPOUND_CID_{cid}'
        response = urllib.request.urlopen(url)
        savename = f"{output}/{digit}/Conformer3D_COMPOUND_{digit}_{lid}_{cid}.sdf"
        
        # url이 가리키는 주소에 접근해서 해당 자원을 로컬 컴퓨터에 저장하기
        urllib.request.urlretrieve(url, savename)
```
